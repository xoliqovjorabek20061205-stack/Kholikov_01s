```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/museum_visitors.csv", index_col=0)
df.head()
```

```python
sns.lineplot(data=df)
plt.show()
```

```python
plt.figure(figsize=(16,6))
plt.title("Muzeylarga tashrif grafigi")
plt.xlabel("Sana")
plt.ylabel("Odamlar soni")
sns.lineplot(data=df)
plt.show()
```

```python
plt.figure(figsize=(16,6))
plt.title("Muzeylarga tashrif grafigi")
plt.xlabel("Sana")
plt.xticks(rotation=90) # x o'qidagi yozuvlarni 90 gradusga aylantiramiz
plt.ylabel("Odamlar soni")
sns.lineplot(data=df)
plt.show()
```

```python
plt.figure(figsize=(16,6))
plt.title("Muzeylarga tashrif grafigi")
plt.xlabel("Sana")
plt.xticks(rotation=90)
plt.ylabel("Odamlar soni")
sns.lineplot(data=df['America Tropical Interpretive Center'])
plt.show()
```

```python
plt.figure(figsize=(16,6))
plt.title("Muzeylarga tashrif grafigi")
plt.xlabel("Sana")
plt.xticks(rotation=90)
plt.ylabel("Odamlar soni")
sns.lineplot(data=df[['Avila Adobe','Firehouse Museum']])
plt.show()
```
